```python
new_df = df[['Date', 'PM10 (ug/m3)', 'PM2.5 (ug/m3)', 'PM 1 (ug/m3)',
       'O3 (ppb)', 'CO (ppm)', 'NO (ppm)', 'NO2 (ppb)', 'Nox (ppb)',
       'CO2 (ppm)']]

new_df.info()
```

```python
# Distribution of pollutants using histograms
pollutants = ['PM10 (ug/m3)', 'PM2.5 (ug/m3)', 'PM 1 (ug/m3)',
       'O3 (ppb)', 'CO (ppm)', 'NO (ppm)', 'NO2 (ppb)', 'Nox (ppb)',
       'CO2 (ppm)']
for pollutant in pollutants:
    plt.figure(figsize=(16, 12))
    sns.histplot(df[pollutant], bins=30, kde=True)
    plt.title(f'Distribution of {pollutant}')
    plt.xlabel(pollutant)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Correlation analysis

correlation_matrix = df.corr()
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Pollutants')
plt.show()
```

```python
# Scatter plot of PM2.5 vs PM10
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PM10 (ug/m3)', y='PM2.5 (ug/m3)', data=df)
plt.title('Scatter plot of PM10 vs PM2.5')
plt.xlabel('PM10 (ug/m3)')
plt.ylabel('PM2.5 (ug/m3)')
plt.show()
```

```python
plt.figure(figsize=(24, 12))
for col in df.columns[2:]:
    plt.plot(df['Date'], new_df[col], label=col)
plt.title('Time Series Plot of Pollutants')
plt.xlabel('Date')
plt.ylabel('Pollution Level')
plt.legend()
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize=(24, 16))
for i, col in enumerate(new_df.columns[1:]):
    plt.subplot(3, 3, i+1)
    sns.kdeplot(df[col], shade=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(24, 12))
sns.boxenplot(data=new_df.drop(columns=['Date']), color="skyblue")
plt.title('Boxen Plot of Pollutants')
plt.ylabel('Concentration')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.jointplot(x='PM2.5 (ug/m3)', y='PM10 (ug/m3)', data=new_df, kind='kde', cmap='viridis')
plt.title('KDE Joint Plot of PM2.5 vs PM10')
plt.tight_layout()
plt.show()
```

```python
# Extract hour and day of the week
new_df['Hour'] = new_df['Date'].dt.hour
new_df['DayOfWeek'] = new_df['Date'].dt.day_name()

# Group by hour and day of the week and calculate mean pollutant levels
hourly_mean = new_df.groupby('Hour').mean().reset_index()
dayofweek_mean = new_df.groupby('DayOfWeek').mean().reset_index()
```

```python
# Plot average pollutant levels by hour of the day
plt.figure(figsize=(24, 12))
for col in new_df.columns[1:-2]:  # Exclude non-pollutant columns
    plt.plot(hourly_mean['Hour'], hourly_mean[col], label=col)
plt.title('Average Pollutant Levels by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Concentration')
plt.legend()
plt.grid(True)
plt.xticks(np.arange(0, 24, step=1))
plt.tight_layout()
plt.show()
```

```python
# Plot average pollutant levels by day of the week
plt.figure(figsize=(24,12))
for col in new_df.columns[1:-2]:  # Exclude non-pollutant columns
    plt.plot(dayofweek_mean['DayOfWeek'], dayofweek_mean[col], label=col)
plt.title('Average Pollutant Levels by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Concentration')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Extract month and year
new_df['Month'] = new_df['Date'].dt.month
new_df['Year'] = new_df['Date'].dt.year

# Group by month and year and calculate mean pollutant levels
monthly_mean = df.groupby(['Year', 'Month']).mean().reset_index()

```

```python
# Create a new 'Date' column from 'Year' and 'Month' for plotting
monthly_mean['Date'] = pd.to_datetime(monthly_mean[['Year', 'Month']].assign(day=1))
```

```python
monthly_mean
```

```python
# Plot average pollutant levels by month and year
plt.figure(figsize=(24, 12))
for col in new_df.columns[1:-4]:  # Exclude non-pollutant columns
    plt.plot(monthly_mean['Date'], monthly_mean[col], label=col)
plt.title('Average Pollutant Levels by Month and Year')
plt.xlabel('Date')
plt.ylabel('Average Concentration')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Extract season
def get_season(month):
    if month in [1,2,3]:
        return 'Spring'
    elif month in [4,5,6]:
        return 'Summer'
    elif month in [7,8,9,10]:
        return 'Autumn'
    else:
        return 'Winter'

new_df['Season'] = new_df['Date'].dt.month.apply(get_season)
```

```python
new_df
```

```python
# Plot boxplots of pollutants by season
plt.figure(figsize=(24, 16))
for i, col in enumerate(new_df.columns[1:-5]):  # Exclude non-pollutant columns
    plt.subplot(3, 3, i+1)
    sns.boxplot(x='Season', y=col, data=new_df)
    plt.title(col)
    plt.xlabel('Season')
    plt.ylabel('Concentration')
plt.tight_layout()
plt.show()
```

```python
# Plot the trend of pollutant levels over different years
plt.figure(figsize=(24, 16))
for col in df.columns[2:11]:  # Exclude non-pollutant and 'Year' columns
    sns.lineplot(data=new_df, x=new_df['Year'], y=col, ci='sd', label=col)
plt.title('Trend of Pollutant Levels Over Different Years')
plt.xlabel('Year')
plt.ylabel('Pollutant Level')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a faceted scatter plot
sns.relplot(data=new_df, x='CO (ppm)', y='O3 (ppb)', col='Hour', col_wrap=4, hue='Hour', palette='viridis', kind='scatter', facet_kws=dict(sharex=False, sharey=False))
plt.suptitle('Relationship Between CO and O3 Across Different Hours of the Day', y=1.02)
plt.show()
```
